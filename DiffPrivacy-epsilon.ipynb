```
import numpy as np
```

```
num_teachers = 10
num_examples = 10000
num_labels = 10
```

```
preds = (np.random.rand(num_examples,num_teachers) * num_labels).astype(int).transpose(0,1)
```

```
len(preds)
```

```

new_labels = []
for an_image in preds:
  label_counts = np.bincount(an_image,minlength=num_labels)
  epsilon = 0.1
  beta = 1/epsilon
  for j in range(len(label_counts)):
    label_counts[j] += np.random.laplace(0,beta,1)
  new_label = np.argmax(label_counts)
  new_labels.append(new_label)
```
